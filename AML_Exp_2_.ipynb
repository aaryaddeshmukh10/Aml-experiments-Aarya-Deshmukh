**Import Libraries and Load Dataset**

```python
import pandas as pd

from sklearn.datasets import load_breast_cancer

from sklearn.feature_selection import (
    SelectKBest,
    chi2,
    mutual_info_classif,
    f_classif,
    VarianceThreshold
)

from sklearn.preprocessing import MinMaxScaler

# Load dataset
data = load_breast_cancer()

df = pd.DataFrame(
    data.data,
    columns=data.feature_names
)

df['Diagnosis'] = data.target

print(df.head())
```

**Separate Features and Target**

```python
X = df.drop('Diagnosis', axis=1)

y = df['Diagnosis']

print("Shape of X:", X.shape)

print("Shape of y:", y.shape)
```

**Correlation Method**

```python
correlation = X.corrwith(y).abs().sort_values(
    ascending=False
)

print("Top 5 Features (Correlation):")

print(correlation.head(5))
```

**Chi-Square Feature Selection**

```python
# Scaling for chi-square
scaler = MinMaxScaler()

X_scaled = scaler.fit_transform(X)

chi2_selector = SelectKBest(
    score_func=chi2,
    k=5
)

chi2_selector.fit(X_scaled, y)

chi2_features = X.columns[
    chi2_selector.get_support()
]

print("Top 5 Features (Chi-Square):")

print(chi2_features)
```

**Mutual Information Method**

```python
mi = mutual_info_classif(X, y)

mi_series = pd.Series(
    mi,
    index=X.columns
).sort_values(ascending=False)

print("Top 5 Features (Mutual Information):")

print(mi_series.head(5))
```

**ANOVA F-Test Method**

```python
f_values, _ = f_classif(X, y)

f_series = pd.Series(
    f_values,
    index=X.columns
).sort_values(ascending=False)

print("Top 5 Features (ANOVA F-Test):")

print(f_series.head(5))
```

**Variance Threshold Method**

```python
selector = VarianceThreshold(
    threshold=0.01
)

selector.fit(X)

var_features = X.columns[
    selector.get_support()
]

print("Top Features (Variance Threshold):")

print(var_features[:5])
```
